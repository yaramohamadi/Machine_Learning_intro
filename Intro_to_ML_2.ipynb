# K-Nearest Neighbors

Let's learn the K-Nearest Neighbors algorithm with an example. We will use the **Iris flower dataset**. <br>
For demonstration, we will use only 2 fields for classification, **Sepal Length**, and **Sepal Width**. Their scatterplot looks like this:


<img src="knn111.png" alt="drawing" width="400"/>


If we have a new flower sample (The white circles), we can check which historical sample is it closest to, and assign the label of that sample to it! 
<br> This is **Nearest Neighbor (1-NN)** classification.

However, judging the label based on the first nearest neighbor might be a poor judgment. Especially when the data point is **around the decision boundaries** or it is an **outlier**.

We can choose **5 nearest neighbors (5-NN)** instead and do a majority vote to determine the class of our new datapoint. This would look something like this:

<img src="knn11.png" alt="drawing" width="400"/>


### Formal Definition of K-NN:

A method for classifying cases based on their similarity to other cases. The distance between 2 points is a measure of their dissimilarity.

Dissimilarity can be expressed with **Euclidean distance**, **Cosine Similarity**, or other metrics depending on the domain and dataset.

### K-NN Algorithm:

- Pick a value for K.
- Calculate the distance of unknown cases from all cases. (High computation and memory demand)
- Select the K-observations in the training data that are nearest to the unknown case.
- Predict the class of the unknown case using the most popular class from the K-nearest neighbors.


### How to calculate similarity?

Euqlidean distance can be expressed as: 

$Dis(x1, x2)=\sqrt{\sum_{i=0}^{n}(x_{1i}-x_{2i})^2}$

E.g. we want to calculate the distance between the new unknown sample and the first sample in the train set:

<img src="iris%203.png" alt="drawing" width="400"/>


$Dis(x_3, x_0)=\sqrt{(5.2-5.1)^2+(3.2-3.5)^2+(2.0-1.4)^2+(0.3-0.2)^2}$=...

Don't forget that dissimilarity measure is highly dependent on data type and domain.



### How to select K?

- **Small K (K=1):** Capturing the noise or anomaly of the data (Overfitting) Not generalized enough to predict out-of-sample data.

- **Large K (K=20):** Losing Information (Underfitting) Overly generalized and not accurate.

We can calculate accuracy on the test set for each K and choose the one that gives the highest test accuracy.

KNN can be used for **regression** as well. We simply need to average the continuous target value for the nearest neighbors.

# Evaluation Metrics in Classification

Just like in Regression, in Classification, model evaluation is done on the test set by comparing the predicted target values $\hat{Y}$ with the actual target values $Y$.

Let's imagine we have a classifier for predicting passengers as COVID positive and negative in an airport. There are 4 types of outcomes:

- **True Positive (TP)**
<br> A passenger who is classified as COVID **positive** and is actually **positive**
- **True Negative (TN)**
<br> A passenger who is classified as COVID **negative** and is actually **negative**

- **False Positive (FP)**
<br> A passenger who is classified as COVID **positive** but is actually **negative**

- **False Negative (FN)**
<br> A passenger who is classified as COVID **negative** but is actually **positive**

### Confusion Matrix

The mentioned outcomes can be plotted with a Confusion Matrix. Plotting the Confusion Matrix of our predictions helps us to further interpret our model's performance. We can analyze these outcomes and see for which class our model is performing poorly or well. 
<br>
For our binary classification problem, we have a 2 by 2 Confusion Matrix as shown here:

<img src="confusion%20matrix3.png" alt="drawing" width="400"/>


General Accuracy of a model can be defined as the percentage of the number of correctly predicted labels among the test set:

$Accuracy = \frac{TP+TN}{TP+FP+TN+FN}=\frac{6+16}{6+16+4+4}=0.7$

Not bad right? Well, we have to think about it twice. We care more about correctly identifying the people who are Covid Positive (TP) and care less about identifying Covid Negative people as Positive (FP).

Accuracy can be **misleading** when **dataset is imbalanced**, or when **predicting a specific class correctly matters more**. We have to use other metrics such as Precision, Recall, and F1-Score.

### Precision vs. Recall

Precision tells us how many of the correctly predicted cases actually turned out to be positive.

$Precision=\frac{TP}{TP+FP}$

This determines whether our model is reliable or not.

Recall tells us how many of the actual positive cases we were able to predict correctly with our model.

$Recall=\frac{TP}{TP+FN}$

<hr>

In our Covid Example, for Covid Positive we have:

$precision = 0.56, Recall = 0.50$

This means that 56\% of the Covid Positive predictions turned out to be Covid Positive. Whereas 50\% of the Covid positive cases were successfully predicted Positive by our model. We need a model with more Recall on Covid Positives!!

In practice, increasing Precision lowers Recall and vise-versa. Sometimes there is no clear distinction between whether Precision is more important or Recall. The F1-score captures both the trends in a single value!

### F1-Score

$F1-score=\frac{2}{\frac{1}{Recall}+\frac{1}{Precision}}$

F1-score is maximum when Precision is equal to Recall. Interpretability of the F1-score is poor. But when combined with other evaluation metrics, it can be a good indicator of how well our model is performing.

Furthermore, plots like the AUC-ROC curve can be useful for model evaluation. Please search it out if you are interested!

## KNN Hands-On

### Import libraries

````python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn import preprocessing
%matplotlib inline
````

````python
df = pd.read_csv('iris.csv')
df.head()
````

````python
import seaborn as sns
sns.pairplot(df.iloc[:, 1:6], diag_kind='hist')
````

### Load data

Dataset is evenly distributed among different classes

````python
df['Species'].value_counts()
````

````python
X = df.iloc[:, 1:5]
y = df['Species'].values
X[:5]
````

### Train/Test Split

````python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=4)
print('Train set:', X_train.shape,  y_train.shape)
print('Test set:', X_test.shape,  y_test.shape)
````

### Normalize Data

Now each feature has zero mean and unit variance.

Note that we should not include the test set in fit function because we are not supposed to be aware of the test data.

````python
scaler = preprocessing.StandardScaler()
X_train = scaler.fit(X_train).transform(X_train.astype(float))
X_train[0:5]
````

````python
X_test = scaler.transform(X_test.astype(float))
X_train[0:5]
````

### Classification with KNN

````python
from sklearn.neighbors import KNeighborsClassifier

k = 4
#Train Model and Predict  
neigh = KNeighborsClassifier(n_neighbors = k).fit(X_train,y_train)
neigh
````

### Prediction

````python
yhat = neigh.predict(X_test)
yhat[0:5]
````

### Evaluation

````python
from sklearn import metrics
print("Train set Accuracy: ", metrics.accuracy_score(y_train, neigh.predict(X_train)))
print("Test set Accuracy: ", metrics.accuracy_score(y_test, yhat))
````

### Testing different Ks

````python
Ks = 30
mean_acc = np.zeros((Ks-1))
std_acc = np.zeros((Ks-1))

for n in range(1,Ks):
    
    #Train Model and Predict  
    neigh = KNeighborsClassifier(n_neighbors = n).fit(X_train,y_train)
    yhat=neigh.predict(X_test)
    mean_acc[n-1] = metrics.accuracy_score(y_test, yhat)

    
    std_acc[n-1]=np.std(yhat==y_test)/np.sqrt(yhat.shape[0])

mean_acc
````

````python
plt.plot(range(1,Ks),mean_acc,'g')
plt.fill_between(range(1,Ks),mean_acc - 1 * std_acc,mean_acc + 1 * std_acc, alpha=0.10)
plt.fill_between(range(1,Ks),mean_acc - 3 * std_acc,mean_acc + 3 * std_acc, alpha=0.10,color="green")
plt.legend(('Accuracy ', '+/- 1xstd','+/- 3xstd'))
plt.ylabel('Accuracy ')
plt.xlabel('Number of Neighbors (K)')
plt.tight_layout()
plt.show()
````

````python
print( "The best accuracy was with", mean_acc.max(), "with k=", mean_acc.argmax()+1) 
````

### Cross-Validation with standardization and multiple scores

````python
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import cross_validate

scores = ['accuracy', 'precision_weighted', 'recall_weighted']

clf = make_pipeline(preprocessing.StandardScaler(), KNeighborsClassifier(n_neighbors = 4))
cross_validate(clf, X, y, cv=4, scoring=scores)
````

# Decision Trees

Let's learn Decision Trees with an example. This is a dataset for prescribing drugs to patients: There are 2 types of drugs, A and B. We want to predict which drug is suitable for a patient. So this is a binary classification problem.



We can build a tree like this and to do desicions (predictions) based on it:

<img src="decision%20tree.png" alt="drawing" width="400"/>


In a decision tree, each node contains an attribute that branches out to lower nodes based on the result of a test on the value of the attribute. The leaf nodes represent our prediction.


### Desicion Tree Algorithm:

- Calculate the significance of each attribute in spliting of data.
- Split data based on the value of the best attribute.
- Go to step 1 unless all data in node belongs to 1 class or until reaching max depth.

### How to measure significance?

The algorithm chooses the most predictive attribute to split the data on. The most predictive is one that by splitting it, we can achieve higher accuracy than other attributes. In other words, predictiveness is based on the increase in the **purity** of nodes. 

E.g. splitting first by Sex does not increase purity in the new nodes. So it's not a good split! Instead, we first split by Age, as shown in the first decision tree we saw. 


### Entropy 

Impurity can be measured by the **Entropy** formula.

**Entropy** is a measure of randomness or uncertainty in data. In a decision tree, it is a measure of chaos within a node. A node with all classes equally present in it has an entropy of 1. A node with only one class present has 0 entropy.

$Entropy=-P(A)\log(P(A))-P(B)\log(P(B))$

where $P(A)$ is the proportion of class A in data.

We choose an attribute that gives us maximum **Information Gain**. This can easily be calculated by having the entropy of the parent and the child nodes.

**Information Gain** = Entropy of ParentNode - Weighted average of Entropy of ChildNodes 


**Why Decision Trees**
- High interpretability
- No preprocessing needed
- Data robustness

### Random Forest

Decision trees are prone to overfitting. We can combine many decision trees and form a decison forest. These algorithms are quite powerful! Details about random forests are out of scope for our lesson.

# Logistic Regression

Logistic Regression is analogous to linear regression but tries to predict a categorical or discrete target instead of a numeric one.

For example, imagine we want to predict whether cancer is Benign or Malignant given its radius. The scatter plot along with the logistic regression fit line looks like this:

<img src="logistic2.png" alt="drawing" width="700"/>


In logistic regression, independent variables should be continuous. If they are categorical, they should be encoded. Logistic regression can be used for multi-class classification as well, but we will stick with binary here.

In logistic, **not only we predict the class of a case**, but we **also predict the probability of the case belonging to each class**. (We then map the probability to the appropriate discrete class)


##### Applications of logistic regression:

- Predict the probability of a person having a heart attack
- Predict the mortality in injured patients
- Predicting a customer's churn
- Predicting the probability of failure of a product
- Predicting the likelihood of a homeowner defaulting on a mortgage

##### When to use logistic regression?:

- When the target value is categorical or binary
- When we need probabilistic results
- When data is linearly separable
- When we need to understand the impact of a feature
<br> We can select the best features based on the statistical significance of the coefficients
(After finding the parameters, a feature $X$ with weight $\Theta$ close to zero has a smaller effect on the prediction than features with large values of $\Theta$

<img src="classification.png" alt="drawing" width="400"/>

E.g. in the figure above, we can obtain a linear decision boundary like this:

$ \Theta_0 + \Theta_1X_1 + \Theta_2X_2 > 0 $

Note that we can also achieve a complex decision boundary using polynomial processing. (Out of scope)


### Formalizing logistic regression

$X \in \mathbb{R}^{m\times n}$

$Y \in {0,1}$

$\hat{Y}=P(Y=1|X)$

$P(Y=0|X)=1-P(Y=1|X)$



### Logistic Regression vs Linear Regression

In this part, we will go over why linear regression cannot be used for binary classification problems and look at **sigmoid function**, the main part of logistic regression.


Mapping the class labels to integer numbers, can we use linear regression to solve this problem?



The line can be expressed as: $\Theta^TX$

We can use linear regression for prediction by passing it through a step function:

$\hat{Y}=
\left\{\begin{matrix}
0 & if &\Theta^TX<0.5\\ 
1 & if &\Theta^TX\geqslant 0.5 \\
\end{matrix}\right.$

$\hat{Y}=Step(\Theta^TX)$

E.g. if $\Theta^TX = 0.3$, then it belongs to class 0.

<img src="step%20function.png" alt="drawing" width="400"/>

##### Problems? 
- We don't know with what probability this observation belongs to class 0?
- Another issue is, the predicted value can become a big number, and there is no difference if $\Theta^TX$ is 1 or 1000. The outcome would still be 1!

We need a smoother function like this one:

$\hat{Y}=\sigma(\Theta^TX)=P(Y=1|X)$

Instead of calculating the $\hat{Y}$ directly, it returns the probability that $\hat{Y}$ is very big or very small. It always returns a value between 0 and 1, depending on how large the $\hat{Y}$ is.

### Sigmoid Function

$\sigma(\Theta^TX)=\frac{1}{1+e^{-\Theta^TX}}$

Notice in the equation that when $\Theta^TX$ gets very big, the exponential term goes towards 0 and the function's output goes towards 1, and when $\Theta^TX$ gets very small, the exponential term goes towards infinity and the function's output goes towards 0.

<img src="sigmoid%20function.png" alt="drawing" width="400"/>

The sigmoid function's output is always between 0 and 1, which makes it proper to interpret the results as probabilities. We can find $\Theta$ through the training process.

### Iterative training of logistic regression

- Initialize $\Theta$.
- Calculate $\hat{Y}=\sigma(\Theta^TX)$ for an observation.
- Compare the output of $\hat{Y}$ with the actual output of the observation, $Y$, and record it as the error.
- Calculate the error for all observations. (Cost function)
- Change the $\Theta$s to reduce the cost.
- Go back to step 2.

A popular algorithm for changing and updating $\Theta$ is gradient descent. We will repeat this process until the accuracy of the model is satisfactory.

### Logistic Regression Training

The main objective of training in logistic regression is to change the parameters of the model until we reach the best estimation of the target values in the dataset.

To understand the training process, we have to look at the cost function first, Then, using the derivative of the cost function, we can find how to change the parameters to reduce the cost or the error:

$Cost(\hat{Y}, Y) = \frac{1}{2}(\sigma(\Theta^TX)-Y)^2$

The square in this equation is used because we don't want negative costs.

We want to reduce the cost overall the dataset:

$J(\Theta)=\frac{1}{m}\sum_{i=1}^{m}Cost(\hat{Y}-Y)$

We can calculate the minimum point of this cost function using the derivative of it,


- Initialize the parameters randomly.
- Feed the cost function with the training set, calculate the error.
- Calculate the gradient of the cost function.
- Update weight with new values by gradient descent.
- Go to step 2 until the cost is small enough.
- Predict the new customer X.

To understand this process better, check up on how gradient descent works.

````python
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import itertools

def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):

    # This function prints and plots the confusion matrix.
    # Normalization can be applied by setting `normalize=True`.
    
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    print(cm)

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    
print(confusion_matrix(y_test, yhat, labels=[1,0]))
````

````python
# Compute confusion matrix
cnf_matrix = confusion_matrix(y_test, yhat, labels=[1,0])
np.set_printoptions(precision=2)

# Plot non-normalized confusion matrix
plt.figure()
plot_confusion_matrix(cnf_matrix, classes=['Benign','Malignant'],normalize= False,  title='Confusion matrix')
````

*   **Recall** is the true positive rate. It is defined as: Recall =  TP / (TP + FN)

Now we are in the position to calculate the F1 scores for each label based on the precision and recall of that label.

The F1 score is the harmonic average of the precision and recall, where an F1 score reaches its best value at 1 (perfect precision and recall) and worst at 0. It is a good way to show that a classifier has a good value for both recall and precision.

# Support Vector Machine (SVM)

SVM is another method used for classification. a Support Vector Machine is a supervised algorithm that can classify cases by finding a separator line, plane, or hyperplane.

### How does it work?

- Mapping data to a high-dimensional feature space
<br> (We can classify data even if data is not linearly separable)

- Finding a separator
<br> (The data should be transformed in such a way that a separator could be drawn as a hyperplane.)

E.g. consider the following figure:

<img src="SVMs.png" alt="drawing" width="400"/>

This is a binary classification problem where the data is not linearly seperable. But by a proper transformation into a higher dimensional space, the boundary between the 2 classes can be defined by a hyperplane. This plane can be used to classify new or unknown cases. 

### How to transfer data in such a way that a seperator could be drawn as a hyperplane?

Imagine that we have a 1-dimensional data that is not linearly separable:
We can transfer it to a 2-dimensional space.

<img src="svm2.png" alt="drawing" width="700"/>

Now the data is linearly separable!

Mapping data to a higher dimensional space is called **kernelling** and the mathematical transformation function is called a **Kernel**.

#### Types of kernel function:

- Linear
- Polynomial
- RBF (Radial Basis Function)
- Sigmoid

Each has its own characteristic, equation, and pros, and cons. You can search all about them if you're interested.

There is no easy way of knowing which function performs best with any data set. So we usually choose different functions in turn and compare the results.

### How to find the best hyperplane seperator after the transformation?

SVM is based on the idea of finding a hyperplane that best divides a dataset into 2 classes as shown here. One reasonable choice as the **best hyperplane** is the one that represents the **largest separation or margin between the 2 classes**.

**Goal:** Choose a hyperplane with the biggest possible margin.

<img src="svm%20margin.png" alt="drawing" width="400"/>



Examples closest to the hyperplane are **support vectors**. It is intuitive that **only support vectors matter for achieving our goal**, and thus, other data points can be ignored. So we try to find the hyperplane in such a way that it has the maximum distance to the support vectors.

Decision Boundary line:
<br>
$W^TX+b=0$

Support vector lines:
<br>
$W^TX+b=1$
<br>
$W^TX+b=-1$

The hyperplane is learned from training data using an **optimization** procedure that **maximizes the margin**. Like many other problems, this optimization problem can also be solved by **gradient descent**.
<br>
The output of the algorithm is the values $W$ and $b$ for the line.
We can then make classifications using this estimated line.
<br>If the equation returns a value **greater than 0**, then the point belongs to the **class which is above the line**, and vice-versa.


The SVM algorithm is out of scope for this course, but there are many good articles and videos about it on the web. please check them out if you are interested!




**SVM Strengths:**
- Accurate in high-dimensional spaces
- Memory efficient (Only need support vector datapoints)

**Pros:**
- Prone to over-fitting -> use soft-margin
- No probability estimation
- Not computationally efficient (Not good for +1k rows)


**SVM is good for classification high dimensional data:**
- Image Recognition
- Text category assignment
- Detecting Spam
- Sentiment Analysis
- Regression, Outlier detection and Clustering
