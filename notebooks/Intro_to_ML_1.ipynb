# Simple Linear Regression

Let's predict Predict Price of all houses using longitude.

A scatter plot clearly shows the relationship between variables where changes in one variable explain or possibly cause changes in the other variable. It also shows that these variables are somewhat linearly related. With linear regression, we can fit a line through the data.




let's assume, the line is a good fit for the data. We can now use it to predict the price of an unknown house:


<hr>

### the Fit line

$\hat{Y} = \Theta_0 + \Theta_1 X_1$

$X_1$ is the independent variable and $\hat{Y}$ is the dependent variable. $\Theta_1$ is the slope or the gradient of the line and $\Theta_0$ is the Intercept. They are also called the coefficients of the linear equation. Linear regression estimates the coefficients of the line.

##### How do we determine if the line is the best fit for the data? 

First, let's assume we've already found the best fit line for our data. Now let's go through all the points and check how well they align with this line.


Best fit means that if we have a longitude $X_1$, its price should be predicted very close to the actual value $Y$. But when we use the fit line, to predict the price, it will return $\hat{Y}$.

The difference between the distance between the data point to the fitted regression line, is also called the residual error:

$Residual Error = |Y - \hat{Y}|$



The mean of all residual errors shows how poorly the line fits with the whole dataset. Mathematically, it can be shown by the Mean Squared Error (MSE) equation:

$MSE = \frac{1}{n}\sum_{i=1}^{n}(Y_i-\hat{Y_i})^2$

So MSE should be minimized. The object of linear regression is to to find the best parameters $\Theta_0$ and $\Theta_1$ that minimize the MSE equation.



##### How to find this line?

- Optimization approach
- Mathematical approach

Lets go with mathematical:

$\hat{Y} = \Theta_0 + \Theta_1 X_1$

Intercept and Slope are calculated using these formulas:

$\Theta_1 = \frac{\sum_{i=1}^{n}(X_i-\bar{X})(Y_i-\bar{Y})}{\sum_{i=1}^{n}(X_i-\bar{X})}$
<br>
$\Theta_0 = \bar{Y} - \Theta_1 \bar{X}$

Sopposing that we are using the first 9 observations from the table below to predict the outcome of the 10th observation:



$\bar{X} = \frac{121.54024 + 121.53951 + ...}{9}=121.53$

$\bar{Y} = \frac{27.9+42+...}{9}=40.36$

$\Theta_1 = ... = 457.897$

$\Theta_0 = 40.36 - (457.897\times 121.53)=-55648.22$

$\Theta_0$ is also called the bias and $\Theta_1$ is the coefficient for the age column.

So the line is: $\hat{Y} = - 55648.22 + 457.897 \times X$

Now making predictions is just plugging the X in the line and finding Y. In our case, The price of a house with $long=121.51486$ is $55.64$.


##### Why linear regression?

Linear regression is:
- Very Fast
- No parameter tuning
- Easy to understand and highly interpretable

## Model Evaluation in Regression Models

After building the model, we have to perform regression evaluation. We must choose an evaluation model that gives us the most accurate results.

How can we calculate the accuracy of our model and how much can we trust this model for the prediction of an unknown sample? <br>One solution for model evaluation is to :
- train on all the dataset, and select a portion of it for testing. This method is ilustrated below:



When evaluating the model, the labels in the test set are not used for prediction. They are only used as ground truth. We pass the feature set of the testing portion to our trained model and predict their target values. Finally, we compare our predicted target values with the ground truth. This indicates how accurate our model is. 

#### The problem with this approach

**Out of sample accuracy:** is the percentage of correct predictions that the model makes on data that it **has not been** trained on. <br>
**Training accuracy** is the percentage of correct predictions that the model makes on data that it **has been** trained on. 
<hr>

## Overfitting

High training accuracy is not necessarily good. Sometimes we over-fit the data when the model is too complex or when we don't have enough data. Overfitting happens that the model is so overly trained to the dataset that it starts to learn the inherent noise in the dataset and produces a non-generalized model. A model that only works well for the data that it has seen before because it has memorized the data instead of learning its valuable patterns.




Remember that the purpose of our model is to make predictions on unknown data! So we care for out-of-sample accuracy

We almost always prefer a high out-of-sample accuracy over a high training accuracy. Doing a train and test on the same dataset can produce a low out-of-sample accuracy due to the likelihood of being over-fit. In order to see how generalized our model is and how well it performs under real unseen data, we must **split train and test** datasets.
<hr> 

## Train/Test Split

In this method, we select a portion of the dataset for training and another portion for testing.



The train and test sets should be mutually exclusive. This is a more accurate evaluation of out-of-sample accuracy because the test set has not been seen before. This setting is more realistic for real-world problems.

Note that in production and deployment, after evaluation, we should ensure that we train our model with the testing set as well as it might contain potentially valuable data.

**The problem with train/test split:** 
<br>
By selecting a different portion of the dataset for training, the accuracy of the model also changes, so we are not able to achieve a fixed accuracy for the model. <br>
To fix this, we can train many models on different portions and simply average their test accuracies!
<br>
This method is called Cross-Validation.

<hr> 

## Cross-Validation

Some Cross-Validation techniques are:

- Leave one out
- k-fold cross-validation
- Time Series cross-validation
- ...

We must select a Cross-Validation method depending on the nature of our task and data. <br>K-fold Cross-Validation is a popular method that can be used in many cases.



### K-Fold

if we have K equals for 4 folds, then we split up the dataset as shown below.



In each fold, a different 25% portion of the dataset is used for testing and the rest is used for training. We calculate accuracy in each setting and finally, the result of all folds are averaged. This gives us a more consistent out-of-sample accuracy.

## Accuracy Metrics for model evaluation

Evaluation metrics, provide a key role in the development of a model. They provide insight into areas that require improvement.

Some commonly used metrics in regression are:

- **Mean Absolute Error**
<br>
$MAE = \frac{1}{n}|Y_i-\hat{Y_i}|$
<br>
Simple average error


- **Mean Squared Error**
<br>
$MSE = \frac{1}{n}\sum_{i=1}^{n}(Y_i-\hat{Y_i})^2$
<br>
More popular since it focuses on large errors


- **Root Mean Squared Error**
<br>
$RMSE = \sqrt{\frac{1}{n}\sum_{i=1}^{n}(Y_i-\hat{Y_i})^2}$
<br>
One of the most popular, since it is interpretable in the same units as Y making it easy to relate its information


- **Relative Absolute Error**
<br>
$RAE = \frac{\sum{|Y_i - \hat{Y_i}|}}{\sum{|y_i-\bar{Y_i}|}}$


- **Relative Squared Error**
<br>
$RSE = \frac{\sum{(Y_i - \hat{Y_i})^2}}{\sum{(y_i-\bar{Y_i})^2}}$


- $R^2$  **(Not an error)**
<br>
A popular metric for the accuracy. Shows how close the data values are to the fitted regression line.

$R^2 = 1 - RSE$

Each metric can be used for quantifying our predictions. The choice of metric completely depends on the type of model, data, and domain of knowledge

# Simple Linear Regression Hands-On

### Importing Packages

````python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
````

### Loading the dataset

````python
df = pd.read_csv("Real estate.csv")

df = df.iloc[:, 2:] # Remove No & transaction date
df.columns = ['age', 'MRT_distance', 'n_stores', 'lat', 'long', 'price']
df.head()
````

### Exploratory Analysis

````python
# How is each column distributed?
df.hist(figsize=(10, 10), grid=False)
plt.show()
````

````python
# Relation of each column with Target (Price)

plt.figure(figsize=(15,3))

for i, col in enumerate(df.columns[:-1]):
    plt.subplot(1,len(df.columns),  i+1)
    plt.scatter(df[col], df['price'],  color='blue')
    plt.xlabel(col)
    if i==0: plt.ylabel("price") 
    else: plt.yticks([])
````

### Creating train and test dataset



````python
#from sklearn.model_selection import train_test_split
#X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

msk = np.random.rand(len(df)) < 0.8
train = df[msk]
test = df[~msk]

len(train), len(test)
````

### Modeling

````python
from sklearn import linear_model
regr = linear_model.LinearRegression()

train_x = np.asanyarray(train[['long']])
train_y = np.asanyarray(train[['price']])

regr.fit(train_x, train_y)

# The coefficients
print ('Coefficients: ', regr.coef_)
print ('Intercept: ',regr.intercept_)
````

In linear regression, **Coefficient** and **Intercept** are the parameters of the fit line. In simple linear regression with only 2 parameters, sklearn can estimate the parameters directly from the data.

### Plot fit line

````python
plt.scatter(train['long'], train['price'],  color='blue', label='trainset')
plt.scatter(test['long'], test['price'],  color='green', label='testset')
plt.plot(train_x, regr.coef_[0][0]*train_x + regr.intercept_[0], '-r')
plt.xlabel("Long")
plt.ylabel("Price")
plt.legend()
````

### Evaluation

````python
from sklearn.metrics import r2_score

test_x = np.asanyarray(test[['long']])
test_y = np.asanyarray(test[['price']])

test_y_ = regr.predict(test_x)

print("Mean absolute error: %.2f" % np.mean(np.absolute(test_y_ - test_y)))
print("Residual sum of squares (MSE): %.2f" % np.mean((test_y_ - test_y) ** 2))
print("R2-score: %.2f" % r2_score(test_y , test_y_) )
````

### Cross-Validation

Cross-Validation using K-fold (K=5) and R^2 scoring. We can perform CV with different score methods. Please check the documentation for more details.

````python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(regr, df[['long']], df[['price']], cv=5, scoring='r2')
scores
````

````python
print("%0.2f R^2 with a standard deviation of %0.2f" % (scores.mean(), scores.std()))
````

# Multiple Linear Regression

Using multiple independent variables to predict the target, is called a Multiple Linear Regression.

use-cases of multiple linear regression:

- To indentify the strength of the effect that the independent variables have on the dependent variable. E.g. does test anxiety, lecturer attendance and gender have any effect on the exam performance of students?

- To understand how the dependent variable changes when we change each independent variable. E.g. Keeping everything else constant, how much does blood pressure go up for every unit increase in the BMI of a patient.

In Multiple linear regression, the dependent variable, $Y$ is a **linear combination** of independent variables, $X$.


$\hat{Y} = \Theta_0 + \Theta_1X_1 + \Theta_2X_2 + ...$

In the case of house price:

$Price = \Theta_0 +\Theta_1Long + \Theta_2Lat + ...$

**In Vector form:**

$\hat{Y} = \Theta^TX$

$\Theta^T = \begin{bmatrix}
\Theta_0 & \Theta_1 & \Theta_2 &...
\end{bmatrix}
$

$X = \begin{bmatrix}
1\\ 
X_1\\ 
X_2\\
...
\end{bmatrix}$

Note that the first X element is set to 0 because it turns $\Theta_0$ into the intercept (or bias parameter).

we call theta the parameters or weight vector of the regression equation.

$\Theta^TX$ in 1 dimensional space is the equation of a line, (simple linear regression). 
<br>
In higher dimensions when we have more than one input, the line is called a plane or a **hyperplane** (Multiple linear regression).

<hr>

### How to find optimized parameters?

Optimized parameters are the ones that lead to a model with the fewest errors.
<br>
Like any regression model, the objective of multiple linear regression is to **minimize MSE**.

How to estimate Theta that minimizes the MSE?

- **Ordinary Least Squares:**
<br>Uses the data as a matrix and uses linear algebra operations to estimate the optimal values for the theta. Has a high time complexity.
<br>Good for small (<10k rows) datasets.

- **Optimization Algorithm:**
<br>Optimizes the coefficients by iteratively minimizing the error of the model on the training data. <br> E.g. Gradient Descent (starts optimization with random values for each coefficient, then calculates the error and tries to minimize it through changing the coefficients in multiple iterations. <br>Good for large datasets.

There are other methods as well that you can explore on your own.
<hr>

In Multiple linear regression, prediction is as easy as plugging in X in the multiple linear regression equation.



##### Some Questions:

- **When to use simple linear regression and when to use multiple linear regression?**
<br>Intuitively, using multiple linear regression results in a better model.

- **How many independent variables should we use? Does adding independent variables always improve the accuracy of the model?**<br>
Adding too many independent variables without any theoretical justification may result in overfitting. an Overfit model is too complicated for the dataset and not general enough to be used for prediction on unseen data.

- **Should independent variables be continuous?**
<br>a categorical independent variable can be incorporated into a regression model by converting it to a numerical variable. E.g. Gender (M, F) can be transformed to (0 and 1). 

**Note** that Multiple linear regression still assumes a linear relationship between each of your independent variables and the dependent variable. If the relationship is non-linear, it is best to use non-linear regression

## Polynomial Regression

Curvy data can be modeled by a polynomial regression of degree $n$.

**A simple polynomial with degree 3:**

$\hat{Y} = \Theta_0 + \Theta_1X + \Theta_2 X^2 + \Theta3 X^3$

Even though the relationship between $X$ and $Y$ is non-linear, a polynomial regression can still be expressed as a linear regression!
<br>
if we declare 3 new independent variables as...

$X_1 = X, $
$X_2 = X^2, $
$X_3 = X^3$

...the model can be converted to a multiple linear regression using these new variables!

$\hat{Y} = \Theta_0 + \Theta_1X_1 + \Theta_2X_2 + \Theta_3X_3$

Now we can treat it like multiple linear regression!

# Polynomial Regression Hands-On

It is evident that **age** and **MRT_distance** have a somewhat curvy relation with the **price**. In this case, we can use polynomial regression. Polynomial Regression models this relationship as an $n_{th}$ degree polynomial function of X. We will try to predict price by a 3 degree polynomial in this example. 




### Polynomial Features


```PolynomialFeatures(degree=n)``` function in sk-learn generates a matrix consisting of all polynomial combinations of the features with degrees less than or equal to the specified degree. 

Now we will generate a polynomial feature set with degree=2 from age and MRT_distance:

````python
from sklearn.preprocessing import PolynomialFeatures
from sklearn import linear_model

train_x = np.asanyarray(train[['MRT_distance', 'age']])
train_y = np.asanyarray(train[['price']])

test_x = np.asanyarray(test[['MRT_distance', 'age']])
test_y = np.asanyarray(test[['price']])

poly = PolynomialFeatures(degree=2)
train_x_poly = poly.fit_transform(train_x)
train_x_poly.shape
````

2 features turned into 6! This is what happened:

$age: A, distance: M$

$$
\begin{bmatrix}
A_1, M_1\\\\
A_2, M_2\\\\
\vdots\\\\
A_n, M_n
\end{bmatrix}\longrightarrow \begin{bmatrix}
1, A_1, A_1^2, M_1, M_1^2, A_1M_1\\\\
1, A_2, A_2^2, M_2, M_2^2, A_2M_2\\\\
\vdots\\\\
1, A_n, A_n^2, M_n, M_n^2, A_nM_n
\end{bmatrix}
$$

Our problem has turned into a multiple linear regression

$$\hat{Y} = b + \Theta_1  A + \Theta_2M + \Theta_3A^2 + \Theta_4A^2 + \Theta_5M^2 + \Theta_6AM$$ 

### Model Fitting

````python
clf = linear_model.LinearRegression()
train_y_ = clf.fit(train_x_poly, train_y)

# The coefficients
print ('Coefficients: ', clf.coef_)
print ('Intercept: ',clf.intercept_)
````

### Evaluation

````python
from sklearn.metrics import r2_score

test_x_poly = poly.transform(test_x)
test_y_ = clf.predict(test_x_poly)

print("MAE: %.2f" % np.mean(np.absolute(test_y_ - test_y)))
print("MSE: %.2f" % np.mean((test_y_ - test_y) ** 2))
print("R2-score: %.2f" % r2_score(test_y,test_y_ ) )
````

You can express the relationship between X and Y by any non-linear function that you can think of!

# Regularization

Ways to reduce overfitting in machine learning models:
- Simplifying model
- Regularization

Sometimes, by reducing the magnitude of the coefficients, we can acheive better results! In other words we can reduce overfitting on the train set and get a more generalized result! This is called Regularization.

**Formal Definition:**

Regularization is a technique used for tuning the function by adding an additional penalty term in the error function. The additional term controls the excessively fluctuating function such that the coefficients don't take extreme values.



In regular regression we minimized this:

$\frac{1}{n}\sum(\hat{Y_i}-Y_i)^2$

There are 2 types of regularization in regression:
- Ridge Regression
<br>$\frac{1}{n}\sum(\hat{Y_i}-Y_i)^2 + \lambda \sum \Theta_j^2$
- Lasso Regression
<br>$\frac{1}{n}\sum(\hat{Y_i}-Y_i)^2 + \lambda \sum |\Theta_j|$

Ridge generally tries to reduce the magnitude of the weights. But never fully discards them, whereas Lasso reduces the magnitude of the weights until some of them actually become zero! Lasso does a sort of feature selection for us.

# Regularization Hands-On

````python
from sklearn.datasets import load_boston
````

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
````

````python
df=load_boston()
df
````

````python
dataset = pd.DataFrame(df.data)
dataset.columns=df.feature_names
dataset["Price"]=df.target
dataset.head()
````

````python
len(dataset)
````

````python
X = dataset.iloc[:, :-1] #independent variables
y = dataset.iloc[:, -1] #dependent variable
````

### Linear Regression

````python
from sklearn.model_selection import cross_validate
from sklearn.linear_model import LinearRegression

reg = LinearRegression()
reg_cv=cross_validate(reg, X, y, scoring='neg_mean_squared_error', cv=5, return_estimator=True)
print(np.mean(reg_cv['test_score']))
````

### Ridge Regression

GridSearch is helpful for searching for best hyperparameters

````python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV

ridge=Ridge()
parameters={'alpha':[1e-15, 1e-10, 1e-8, 1e-3, 1e-2, 1, 5, 10, 20, 30, 35, 40, 45, 50, 55, 100]}
reg_ridge = GridSearchCV(ridge, parameters, scoring='neg_mean_squared_error', cv=5)
reg_ridge.fit(X, y)
````

````python
print(reg_ridge.best_params_)
print(reg_ridge.best_score_)
````

### Lasso Regression

````python
from sklearn.linear_model import Lasso

lasso=Lasso()
parameters={'alpha':[ 1e-15, 1e-10, 1e-8, 1e-3, 1e-2, 1, 5, 10, 20, 30, 35, 40, 45, 50, 55, 100]}
reg_lasso = GridSearchCV(lasso, parameters, scoring='neg_mean_squared_error', cv=5)

reg_lasso.fit(X, y)
print(reg_lasso.best_params_)
print(reg_lasso.best_score_)
````

### Comparison

````python
plt.figure(figsize=(7, 10))

plt.subplot(3, 1, 1)
plt.bar(dataset.columns[:-1],reg_cv['estimator'][2].coef_)
plt.ylim([-25, 10])
plt.xticks([])
plt.title('Linear Regression')
plt.ylabel('Magnitude')

plt.subplot(3, 1, 2)
plt.bar(dataset.columns[:-1],reg_ridge.best_estimator_.coef_)
plt.ylim([-25, 10])
plt.xticks([])
plt.title('Ridge Regression')
plt.ylabel('Magnitude')

plt.subplot(3, 1, 3)
plt.bar(dataset.columns[:-1],reg_lasso.best_estimator_.coef_)
plt.ylim([-25, 10])
plt.title('Lasso Regression')
plt.xlabel('Coefficients')
plt.ylabel('Magnitude')
````
